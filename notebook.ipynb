```python
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams, PointStruct

# Local file-backed (persists between runs)
client = QdrantClient(path="./qdrant_data")

# In-memory (wiped on exit)
# client = QdrantClient(":memory:")

# Create a collection (like a table, but for vectors)
client.create_collection(
    collection_name="components",
    vectors_config=VectorParams(size=768, distance=Distance.COSINE),
)
```

```python
# Insert vectors
client.upsert(
    collection_name="components",
    points=[
        PointStruct(
            id=1,
            vector=[0.1, 0.2, ...],   # your embedding, must match size=768
            payload={"name": "Front Brake Caliper", "subsystem": "Braking"}
        ),
        PointStruct(
            id=2,
            vector=[0.3, 0.1, ...],
            payload={"name": "ABS Modulator Valve", "subsystem": "Braking"}
        ),
    ]
)

# Search
results = client.query_points(
    collection_name="components",
    query=[0.15, 0.18, ...],   # embedding of your query text
    limit=5,
)

for r in results.points:
    print(r.id, r.score, r.payload)
```
